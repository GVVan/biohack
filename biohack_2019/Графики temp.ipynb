```python
import sys
import numpy as np
import glob
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib notebook
```

```python
filenames = glob.glob('training/*.psv')
```

```python
# разбиение на пацинетов с патолгией и без
patient_1 = []
patient_0 = []
for l in filenames:
    data = pd.read_csv(l,delimiter='|')
    if sum(data.SepsisLabel) > 0:
        patient_1.append(l)
    else:
        patient_0.append(l)
```

```python
# поиск пациента для графика температуры
```

```python
for i in patient_1:    
    data = pd.read_csv(i,delimiter='|')
    statinfo = os.stat(i)
    if data.Temp.std() > 1.3:
        print([statinfo.st_size, i, data.Temp.std()])
```

```python
for i in range(len(patient_1)):
    if patient_1[i] == 'training\\p01857.psv':
        ind = i
        print(i)
```

```python
data = pd.read_csv(patient_1[ind],delimiter='|')
marks0 = np.array(data.SepsisLabel[data.SepsisLabel == 0].index)
marks1 = np.array(data.SepsisLabel[data.SepsisLabel == 1].index)
data = np.array(data.Temp.interpolate())
#data = np.array([x for x in data if str(x) != 'nan'])
```

```python
# график пациента с сепсисом
plt.plot(marks0,data[marks0],label="presepsis")
plt.plot(marks1-1,data[marks1-1],label="sepsis")
plt.legend(bbox_to_anchor=(0.95, 0.2), loc=1)#, borderaxespad=0.)
plt.ylabel('Temperature (⁰C)')
plt.xlabel('Hours')

plt.show()
```
