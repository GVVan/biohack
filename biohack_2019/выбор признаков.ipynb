```python
import sys
import numpy as np
import glob
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
filenames = glob.glob('training/*.psv')
```

```python
dict_acc = []
```

```python
# HR
TP = 0
TN = 0
FP = 0
FN = 0
nn = 0

pred = []  
test = []
for l in filenames:
    data = pd.read_csv(l,delimiter='|')
    #data = data.fillna(0)
    if data.HR.count() == 0:
        nn +=1
        continue
    elif len(data.HR[(data.HR > 90) | ((data.HR < 59) & (data.HR == 0))]) > 0 and sum(data.SepsisLabel) > 0:
        TP += 1
        pred.append(1)
        test.append(1)
    elif len(data.HR[(data.HR > 90) | ((data.HR < 59) & (data.HR == 0))]) > 0 and sum(data.SepsisLabel) == 0:
        FP += 1
        pred.append(1)
        test.append(0)
    elif len(data.HR[(data.HR > 90) | ((data.HR < 59) & (data.HR == 0))]) == 0 and sum(data.SepsisLabel) > 0:
        FN += 1
        pred.append(0)
        test.append(1)
    elif len(data.HR[(data.HR > 90) | ((data.HR < 59) & (data.HR == 0))]) == 0 and sum(data.SepsisLabel) == 0:
        TN += 1
        pred.append(0)
        test.append(0)
        
    '''
    elif sum(data.SepsisLabel) > 0:
        analyseNaN.append(1)
    elif sum(data.HR) == 0 and sum(data.SepsisLabel) == 0:
        analyseNaN.append(1)
    '''
SE = TP/(TP+FN)*100
SP = TN/(TN+FP)*100
dict_acc.append(['HR', 'SE ' + str(round(SE,2)) , 'SP ' + str(round(SP,2)), 'nan = ' + str(nn)])

# roc curve
fpr, tpr, threshold = metrics.roc_curve(test, pred)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic, HR')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
# Resp
TP = 0
TN = 0
FP = 0
FN = 0
nn = 0

pred = []  
test = []
for l in filenames:
    data = pd.read_csv(l,delimiter='|')
    #data = data.fillna(0)
    # патология Resp
    if data.Resp.count() == 0:
        nn += 1
        continue
    elif len(data.Resp[(data.Resp > 20)]) > 0 and sum(data.SepsisLabel) > 0:
        TP += 1
        pred.append(1)
        test.append(1)
    elif len(data.Resp[(data.Resp > 20)]) > 0 and sum(data.SepsisLabel) == 0:    
        FP += 1
        pred.append(1)
        test.append(0)
    elif len(data.Resp[(data.Resp > 20)]) == 0 and sum(data.SepsisLabel) > 0:
        FN += 1
        pred.append(0)
        test.append(1)
    elif len(data.Resp[(data.Resp > 20)]) == 0 and sum(data.SepsisLabel) == 0:
        TN += 1
        pred.append(0)
        test.append(0)
        
SE = TP/(TP+FN)*100
SP = TN/(TN+FP)*100
dict_acc.append(['Resp', 'SE ' + str(round(SE,2)) , 'SP ' + str(round(SP,2)), 'nan = ' + str(nn)])

# roc curve
fpr, tpr, threshold = metrics.roc_curve(test, pred)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic, Resp')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
